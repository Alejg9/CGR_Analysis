```python
# Data processing and model train

from libs.ml import AtmosphericalData
from tensorflow import keras
def train_model(target_column:str, date_col: str, train_column: str, n_steps:int, training_args: dict, model_args: list):
    ## Read data and preprocessing

    inside = []
    for v in model_args:
        layer = getattr(keras.layers, v['layer'])

        inside.append(layer(v['units'], **v['args']))

    model = keras.models.Sequential(inside)

    ad = AtmosphericalData(model=model, path=f"./data/{target_column}/{target_column}_decomposition.csv", use_cudnn=False)
    del model

    print(ad.df.columns)

    ad.convert_date(date_col=date_col)
    ad.target_selection(train_column)
    ad.data_preprocessing(n_steps=n_steps, train_vol=0.9)

    ## Train data

    ad.train_data(
        epochs=training_args['epochs'],
        batch_size=training_args['batch_size'],
        validation_split=training_args['validation_split'],
        verbose=training_args['verbose']
    )


    ## Test trained model
    ad.test_model()


    ## Plot comparison
    ad.plot_tests()

    return ad
```

```python
# Batch size to train
n_steps = 12
target = 'ws(mps)'

model_args_r = [
    {'layer': 'Input', 'units': (12,3), 'args': {}},
    {'layer': 'LSTM', 'units':12, 'args': {'return_sequences': True, 'use_cudnn': False, 'recurrent_dropout': 0.0}},
    {'layer': 'LSTM', 'units':12, 'args': {'return_sequences': True, 'use_cudnn': False, 'recurrent_dropout': 0.0}},
    {'layer': 'LSTM', 'units':12, 'args': {'return_sequences': True, 'use_cudnn': False, 'recurrent_dropout': 0.0}},
    {'layer': 'LSTM', 'units':32, 'args': {'return_sequences': False, 'use_cudnn': False, 'recurrent_dropout': 0.0}},
#    {'layer': 'Dense', 'units': 6, 'args': {}},
    {'layer': 'Dense', 'units': 1, 'args': {}},
]

training_args_r = {
    'epochs':1000,
    'batch_size':36,
    'validation_split':0.1,
    'verbose':1
}

model_args = [
    {'layer': 'InputLayer', 'units': (n_steps,3), 'args': {}},
    {'layer': 'LSTM', 'units':240, 'args': {'return_sequences': True, 'use_cudnn': False, 'recurrent_dropout': 0.0}},
    {'layer': 'LSTM', 'units':480, 'args': {'return_sequences': False, 'use_cudnn': False, 'recurrent_dropout': 0.0}},
    {'layer': 'Dense', 'units': 120, 'args': {}},
    {'layer': 'Dense', 'units': 1, 'args': {}},
]
training_args = {
    'epochs':300,
    'batch_size':96,
    'validation_split':0.05,
    'verbose':0
}

ad_r = train_model(target_column=target, date_col='date', train_column='resid', n_steps=n_steps, training_args=training_args_r, model_args=model_args_r)
ad_t = train_model(target_column=target, date_col='date', train_column='trend', n_steps=n_steps, training_args=training_args, model_args=model_args)
ad_s = train_model(target_column=target, date_col='date', train_column='seasonal', n_steps=n_steps, training_args=training_args, model_args=model_args)
ad_o = train_model(target_column=target, date_col='date', train_column='observed', n_steps=n_steps, training_args=training_args, model_args=model_args)

```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

true_rescaled = ad_r._true_rescaled + ad_t._true_rescaled + ad_s._true_rescaled
pred_rescaled = ad_r._pred_rescaled + ad_t._pred_rescaled + ad_s._pred_rescaled

rmse = np.sqrt(mean_squared_error(true_rescaled, pred_rescaled))
mae = mean_absolute_error(true_rescaled, pred_rescaled)
print("%RMSE:", rmse * 100)
print("%MAE:", mae * 100)
plt.figure(figsize=(14,5))

plt.plot(ad_r.df[ad_r._date_col], ad_r.ds + ad_t.ds + ad_s.ds, label="Real", alpha=0.5)
plt.plot(ad_r._test_dates, ad_r._pred_rescaled + ad_s._pred_rescaled + ad_t._pred_rescaled, label="Predicción", color="red", linewidth=2)
plt.title("Predicción LSTM de " + ad_r._target_col)
plt.xlabel(ad_r._date_col)
plt.ylabel(ad_r._target_col)
plt.legend()
plt.grid()
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
def build_dates(year):
    return [f'{year}{month}01' if month >= 10 else f'{year}0{month}01' for month in range(1, 13)]
dates_input = build_dates(2023)
dates_input = dates_input + build_dates(2024)
predicts = []
predicts1 = []
for date in dates_input:
    _predict = ad_r.predict_for_date(date)
    _predict += ad_t.predict_for_date(date)
    _predict += ad_s.predict_for_date(date)
    predicts.append(_predict)

    _predict = ad_o.predict_for_date(date)
    predicts1.append(_predict)


plt.figure(figsize=(14,5))
plt.plot(pd.to_datetime(dates_input, format='%Y%m%d'), predicts, label="Predict S+T+R", color='white', linewidth=2)
plt.plot(pd.to_datetime(dates_input, format='%Y%m%d'), predicts1, label="Predict Obs", color='green', linewidth=1)
plt.plot(ad_r.df['date'], ad_r.df['observed'], label="Real", color="red", linewidth=2)
plt.title("Predicción LSTM de " + target)
plt.xlabel("Fecha")
plt.ylabel(target)
plt.legend()
plt.grid()
plt.tight_layout()
```
