```python
import tensorflow as tf
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    # Restrict TensorFlow to only allocate 1GB of memory on the first GPU
    try:
        tf.config.set_logical_device_configuration(
            gpus[0],
            [tf.config.LogicalDeviceConfiguration(memory_limit=6*1024)])
        logical_gpus = tf.config.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        # Virtual devices must be set before GPUs have been initialized
        print(e)
```

```python
from libs.ml import AtmosphericalData
from tensorflow import keras
# Batch size to train
n_steps = 48



#model = tf.keras.Sequential([
#    tf.keras.layers.LSTM(
#        64, 
#        return_sequences=True, 
#        activation='tanh', 
#        recurrent_dropout=0.3, 
#        input_shape=(n_steps, 3), 
#        use_cudnn=False),
#    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32, return_sequences=True, use_cudnn=False)),
#    tf.keras.layers.Dense(32, activation='tanh'),
#    tf.keras.layers.Dense(1, activation='relu')  # Non-negative output
#])


model = keras.models.Sequential([
    keras.layers.LSTM(240, return_sequences=True, input_shape=(n_steps, 3), use_cudnn=False, recurrent_dropout=0.0),
    keras.layers.LSTM(240, return_sequences=True, input_shape=(n_steps, 3), use_cudnn=False, recurrent_dropout=0.0),
    keras.layers.LSTM(360, return_sequences=True, input_shape=(n_steps, 3), use_cudnn=False, recurrent_dropout=0.0),
    keras.layers.LSTM(480, return_sequences=False, input_shape=(n_steps, 3), use_cudnn=False , recurrent_dropout=0.0),
    keras.layers.Dense(120, activation='tanh'),
    keras.layers.Dense(1)
])

# Data processing and model train

## Read data and preprocessing
ad = AtmosphericalData(model=model, path="./data/CGR.xlsx", use_cudnn=False)
del model
target_column = 'ws(mps)'
print(ad.df.columns)

ad.convert_date()
ad.target_selection(target_column)
ad.data_preprocessing(n_steps=n_steps, train_vol=0.95)

## Train data
epochs=300
batch_size=96
validation_split=0.05
verbose=1
ad.train_data(
    epochs=epochs,
    batch_size=batch_size,
    validation_split=validation_split,
    verbose=verbose
)


## Test trained model
ad.test_model()


## Plot comparison
ad.plot_tests()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
def build_dates(year):
    return [f'{year}{month}01' if month >= 10 else f'{year}0{month}01' for month in range(1, 13)]
dates_input = build_dates(2023)
dates_input = dates_input + build_dates(2024)
predicts = []

for date in dates_input:
    _predict = ad.predict_for_date(date)
    predicts.append(_predict)


plt.figure(figsize=(14,5))
plt.plot(pd.to_datetime(dates_input, format='%Y%m%d'), predicts, label="Predict", alpha=0.5)
plt.plot(ad.df['Fecha'], ad.df[target_column], label="Real", color="red", linewidth=2)
plt.title("Predicción LSTM de " + target_column)
plt.xlabel("Fecha")
plt.ylabel(target_column)
plt.legend()
plt.grid()
plt.tight_layout()
```
